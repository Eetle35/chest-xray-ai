```python
from pathlib import Path
import matplotlib.pyplot as plt
import random
from PIL import Image

# Root directory for all dataset splits
DATA_DIR = Path("/home/rishi/medai/data/chest-xray-pneumonia/chest-pneumonia/archive/chest_xray/chest_xray")

# Print image counts per split and class to audit the dataset
for split in ["train", "val", "test"]:
    for cls in ["NORMAL", "PNEUMONIA"]:
        count = len(list((DATA_DIR / split / cls).glob("*.jpeg")))
        print(f"{split}/{cls}: {count}")
```

```python
fig, axes = plt.subplots(2, 4, figsize=(12, 6))

# Show 4 random samples from each class to visually inspect image quality
for row, cls in enumerate(["NORMAL", "PNEUMONIA"]):
    images = list((DATA_DIR / "train" / cls).glob("*.jpeg"))
    samples = random.sample(images, 4)
    for col, img_path in enumerate(samples):
        img = Image.open(img_path)
        axes[row, col].imshow(img, cmap="gray")
        axes[row, col].set_title(cls)
        axes[row, col].axis("off")

plt.tight_layout()
plt.savefig("../outputs/sample_images.png")
plt.show()
```
